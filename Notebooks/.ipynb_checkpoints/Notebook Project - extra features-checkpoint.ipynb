```python
train = np.array(data)
test = np.array(testdata)

train_size = train.shape[0]
test_size = test.shape[0]
```

## Adding (useable) feature for first_active_month
Here the first_active_month is transformed by multiplying the years by 12 and adding the months. The "magic" number 24143 is the very first month a card was recorded to be first active: 2011-11, November 2011, so that is substracted to let the feature's minimum be 0.

```python
# for training set
train_fam_feature = np.zeros(train_size)

for i in range(train_size):
    feature_row = train[i,0].split('-')
    train_fam_feature[i] = int(feature_row[0])*12 + int(feature_row[1]) - 24143
    
# for test set
test_fam_feature = np.zeros(test_size)

for i in range(test_size):
    
    if type(test[i,0]) != float: #if-statement, because missing value
        feature_row = test[i,0].split('-')
    test_fam_feature[i] = int(feature_row[0])*12 + int(feature_row[1]) - 24143
```

```python
# inserting the feature in training data
data.insert(5,"fam_feature",train_fam_feature)

# inserting the feature in test data
testdata.insert(5,"fam_feature",test_fam_feature)
```

## Historical transactions - feature for city_id & state_id
A feature will be extracted for the city_id. This feature should represent how locally the customer buys. The formula for this is: (_transactions at most visited city_ / _total transactions_) * (_transactions at most visited state_ / _total transactions_).
The feature is a value between 0 and 1. If the feature returns 1, then the customer buys only locally. If the feature returns a number closer to 0, then the customer buys frequently outside his city and/or state.

```python
historical_transactions = pd.read_csv('all/historical_transactions.csv')
```

```python
historical_transactions
```

```python
MVC = 0
MVS = 0
transactions = 0
start = 0

for i in range(100):
    MVC = historical_transactions.loc[historical_transactions['card_id'] == train[i+start,1]].city_id.value_counts().max()
    MVS = historical_transactions.loc[historical_transactions['card_id'] == train[i+start,1]].state_id.value_counts().max()
    transactions = historical_transactions.loc[historical_transactions['card_id'] == train[i+start,1]].city_id.count()
    print( (MVC*MVS)/transactions**2 )
```
